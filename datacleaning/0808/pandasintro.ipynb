```python
import numpy as np
import pandas as pd
```

```python
series1 = pd.Series([10, 20, 30, 40])  # 인덱스 없이 0, 1, 2, 3으로 접근하도록 생성
series2 = pd.Series([10, 20, 30, 40], index=["one", "two", "three", "four"])
print(series1)
print(series2)
series3 = pd.Series({"이름":"kim", "나이":"24", "직업":"학생"})  # key -> index, value -> value
print(series3)
```

```python
# 인덱싱
print(series3[0])  # 일련번호로 접근
print(series3["나이"])  # 인덱스로 접근
print(series3[0:2])  # 일련번호로 접근하면 종료 번호 앞까지
print(series3["이름":"직업"])  # 인덱스로 접근하면 종료인덱스까지
```

```python
print(series1.values)  # value 추출
print(series1 + 100)  # value로 연산 수행
print(np.sum(series1))
```

```python
s1 = pd.Series({"한국":8, "미국":6, "일본":5, "중국":3 })
s2 = pd.Series({"한국":np.nan, "미국":6, "일본":5, "독일":3 })
print(s1+s2)
```

```python
des = {
    'title' : ['어린왕자', '페르마의 마지막 정리', '사미인곡'],
    'author' : ['생텍쥐페리', '사이먼싱', '정철'],
    'price' : [10000, 20000, 30000]
}

df = pd.DataFrame(des)
print(df)
```

```python
print(df.index)
print(df.columns)
df.index = range(1,4)
df.columns = ['제목', '저자', '가격']
print(df)
```

```python
# head, tail 함수는 데이터가 제대로 불러졌는지 확인할 때 사용
print(df.head(1))
print(df.tail(1))
print(type(df.values))  # numpy.ndarray이어야 머신러닝에 사용 가능
# 데이터 분석 전 데이터 정보 확인
print(df.info())
```
