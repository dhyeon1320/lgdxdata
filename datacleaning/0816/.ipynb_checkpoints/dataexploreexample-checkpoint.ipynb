```python
# import libraries
import numpy as np
import pandas as pd

# 시각화 패키지
import matplotlib.pyplot as plt

# 한글 출력 관련 libraries
import platform
from matplotlib import font_manager, rc
```

```python
# read data
cctv = pd.read_excel('../data/data/cctv.xlsx')
print(cctv.head())
cctv.info()
```

```python
# pop.txt - 한글 포함 / 구분 기호 : 탭 / 천 단위 구분 기호 존재
pop = pd.read_csv('../data/data/pop.txt', skiprows=2, delimiter='\t', thousands=',')
print(pop.head())
pop.info()
```

```python
# column 이름 바꾸기
cctv.rename(columns={cctv.columns[0]:'구별'}, inplace=True)
pop.rename(columns={pop.columns[1]:'구별'}, inplace=True)

# cctv 공백 존재, pop 공백 없음
region = []
for x in cctv['구별']:
    region.append(x.replace(' ',''))

    
cctv['구별'] = region

print(region)
print(cctv)
```

```python
# 불필요한 열과 행 제거
# pop 데이터에서 기간, 구별, 계, 남자, 여자 열만 필요
pop = pop[['기간', '구별', '계', '남자', '여자']]

# pop 데이터 첫번째 행은 전체 데이터 집계라서 불필요
pop.drop([0], inplace=True)

# 새로운 열 추가 - 없는 열 이름에 데이터 대입
pop['여성비율'] = pop['여자']/pop['계']*100

print(pop)
```

```python
df = pd.merge(cctv, pop, on='구별')
print(df)
```

```python
# 불필요한 데이터 삭제
# del 이용 - pandas에서는 비추천
del df['2011년 이전']
del df['2012년']
print(df)
```

```python
# index 설정
df.set_index('구별', inplace=True)
print(df)
```

```python
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname = 'c:/Windows/Fonts/malgun.ttf').get_name()
    rc('font', family=font_name)
    
df['소계'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
df['cctv비율'] = df['소계']/df['계']*100
df['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['계'], df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv 개수')
plt.grid()
plt.show()
```

```python
fp1= np.polyfit(df['계'], df['소계'],1)
f1 = np.poly1d(fp1)
print(f1)

# 산점도와 단항식을 같이 출력
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize=(10, 10))
plt.scatter(df['계'], df['소계'], s=50)
plt.plot(fx, f1(fx), ls= 'dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('cctv 개수')
plt.grid()
plt.show()
```

```python
# 오차
df['오차'] = np.abs(df['소계'] - f1(df['계']))


plt.figure(figsize=(10, 10))
plt.scatter(df['계'], df['소계'], c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls= 'dashed', lw=3, color='g')

# 좌표로 지정해서 레이블 출력
for n in range(24):
    plt.text(df['계'][n]*1.02, df['소계'][n]*0.98,)

plt.xlabel('인구수')
plt.ylabel('cctv 개수')
plt.grid()
plt.show()
```

```python
df = pd.read_csv('../data/data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
print(df.head())

# 데이터 자료형은 dtypes, info()로 확인
print(df.dtypes)
df.corr()
```

```python
# df['horsepower'] = df['horsepower'].astype('float')  # horsepower 자료형 숫자로 바꾸려는데 에러 could not convert string to float: '?'
print(df['horsepower'].unique())  # '?' 값 존재

# '?' 값 처리 결정 - ? 값을 None으로 설정하고 None 데이터 제거
df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)

# astype 자료형 변경
df['horsepower'] = df['horsepower'].astype('float')

print(df['horsepower'].unique())
```

```python
# origin - 1,2,3 이지만 미국,유럽,일본 의미 -> 새로운 column 추가하거나 column 값 변경 / 범주형으로 변경하면 분석 속도 높아짐
df['origin'].replace({1:'미국', 2:'유럽', 3:'일본'}, inplace=True)
df['origin'] = df['origin'].astype('category')  # 범주형으로 변환
df.info()
print(df.head())
df.corr()
```
